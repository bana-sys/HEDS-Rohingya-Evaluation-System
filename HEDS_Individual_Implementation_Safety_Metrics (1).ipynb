```python
# Humanitarian Engineering and Data Science: LLM Safety Auditor
# Implementation: Ragas Faithfulness Metrics

# 1. Environment Configuration
# Install essential libraries for data manipulation and NLP evaluation
!pip install -q pandas matplotlib datasets

import pandas as pd
import matplotlib.pyplot as plt
from datasets import Dataset

# 2. Data Ingestion: Remote Repository Access
# Pulling the verified humanitarian dataset from GitHub to ensure source integrity
url = "https://raw.githubusercontent.com/bana-sys/HEDS-Rohingya-Evaluation-System/refs/heads/main/humanitarian_eval_data.csv"
df = pd.read_csv(url)

# 3. Technical Implementation: Safety Audit Logic
# We simulate the Ragas 'Faithfulness' metric (0.0 - 1.0) to quantify AI accuracy.
# 1.0 = High Faithfulness; 0.2 = Intentional Hallucination for testing
mock_scores = [1.0, 1.0, 0.2]
df['faithfulness'] = mock_scores

# 4. Quantitative Analysis: System Performance Visualisation
# Generating a bar chart to identify critical safety failures in crisis communication.
# This provides the necessary quantitative backup required for professional evaluation.
plt.figure(figsize=(10, 6))

# Conditional styling: Green for safe responses, Red for high-risk hallucinations
colors = ['#27ae60' if s == 1.0 else '#c0392b' for s in df['faithfulness']]

plt.bar(['Shelter Alert', 'Food Logistics', 'Water Safety'], df['faithfulness'], color=colors)
plt.axhline(y=1.0, color='blue', linestyle='--', label='Required Safety Threshold')

plt.title('Humanitarian AI Safety Audit: Faithfulness per Response Category')
plt.ylabel('Faithfulness Score (0.0 - 1.0)')
plt.legend()
plt.show()

# 5. Ethical Guardrail: Automated Risk Detection
# Programmatically flagging responses that pose a risk to the community.
print("\n--- SAFETY AUDIT SUMMARY ---")
critical_fails = df[df['faithfulness'] < 1.0]

if not critical_fails.empty:
    print("CRITICAL WARNING: Hallucination detected in life-critical alert:")
    # Outputting the specific query and the unfaithful AI response for human review
    print(critical_fails[['user_query', 'ai_generated_response']])
else:
    print("All alerts meet the required faithfulness threshold.")
```
